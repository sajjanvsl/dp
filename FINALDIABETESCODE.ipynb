```python
# ‚úÖ Imports
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
from imblearn.combine import SMOTEENN
from tensorflow.keras.models import Model, load_model
from tensorflow.keras.layers import Input, Dense, Layer
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
import tensorflow.keras.backend as K
import matplotlib.pyplot as plt
from collections import Counter

# =====================================================
#                 DATASET + PREPROCESSING
# =====================================================

df = pd.read_csv('./24-04-2025-1192ALLSAMPLES.csv')
X = df.drop('class', axis=1)
y = df['class']

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
top_features = pd.Series(rf.feature_importances_, index=X.columns).nlargest(10).index.tolist()
X = df[top_features]

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_scaled, y)

# =====================================================
#                        GAN
# =====================================================

class GANGen(nn.Module):
    def __init__(self, z_dim, x_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(z_dim, 64), nn.ReLU(),
            nn.Linear(64, x_dim), nn.Sigmoid()
        )
    def forward(self, z): return self.model(z)

class GANDisc(nn.Module):
    def __init__(self, x_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(x_dim, 64), nn.ReLU(),
            nn.Linear(64, 1), nn.Sigmoid()
        )
    def forward(self, x): return self.model(x)

def generate_gan(X_class, n_samples=500, z_dim=32, epochs=150):
    device = torch.device("cpu")
    G = GANGen(z_dim, X_class.shape[1]).to(device)
    D = GANDisc(X_class.shape[1]).to(device)
    g_opt = optim.Adam(G.parameters(), lr=0.0002)
    d_opt = optim.Adam(D.parameters(), lr=0.0002)
    loss_fn = nn.BCELoss()

    loader = DataLoader(TensorDataset(torch.tensor(X_class, dtype=torch.float32)),
                        batch_size=64, shuffle=True)

    for _ in range(epochs):
        for real_batch, in loader:
            real_batch = real_batch.to(device)
            bs = real_batch.size(0)

            real_labels = torch.ones(bs, 1).to(device)
            fake_labels = torch.zeros(bs, 1).to(device)

            z = torch.randn(bs, z_dim).to(device)
            fake_batch = G(z)

            d_real = D(real_batch)
            d_fake = D(fake_batch.detach())
            d_loss = loss_fn(d_real, real_labels) + loss_fn(d_fake, fake_labels)
            d_opt.zero_grad(); d_loss.backward(); d_opt.step()

            g_loss = loss_fn(D(fake_batch), real_labels)
            g_opt.zero_grad(); g_loss.backward(); g_opt.step()

    with torch.no_grad():
        z = torch.randn(n_samples, z_dim).to(device)
        return G(z).cpu().numpy()

X_class1 = X_resampled[y_resampled == 1]
X_class0 = X_resampled[y_resampled == 0]
gan_data_1 = generate_gan(X_class1, n_samples=500)
gan_data_0 = generate_gan(X_class0, n_samples=500)

X_final = np.vstack([X_resampled, gan_data_1, gan_data_0])
y_final = np.concatenate([y_resampled,
                          np.ones(len(gan_data_1)),
                          np.zeros(len(gan_data_0))]).astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, stratify=y_final, random_state=42
)

# =====================================================
#                   CUSTOM RBF LAYER
# =====================================================

class RBFLayer(Layer):
    def __init__(self, units, gamma=None, **kwargs):
        super().__init__(**kwargs)
        self.units = units
        self.gamma = gamma

    def build(self, input_shape):
        self.centers = self.add_weight(
            name='centers',
            shape=(self.units, int(input_shape[1])),
            initializer='uniform',
            trainable=True
        )
        if self.gamma is None:
            self.gamma = 1.0
        self.built = True

    def call(self, inputs):
        C = K.expand_dims(self.centers, axis=0)
        X = K.expand_dims(inputs, axis=1)
        return K.exp(-self.gamma * K.sum(K.square(X - C), axis=-1))

# =====================================================
#                   BUILD RBFNN MODEL
# =====================================================

def build_rbfnn(input_dim):
    inputs = Input(shape=(input_dim,))
    rbf = RBFLayer(30, gamma=0.5)(inputs)
    output = Dense(1, activation='sigmoid')(rbf)
    model = Model(inputs=inputs, outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy', metrics=['accuracy'])
    return model

model = build_rbfnn(X_train.shape[1])

history = model.fit(
    X_train, y_train, epochs=100, batch_size=16, validation_split=0.1,
    callbacks=[ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)],
    verbose=1
)

# =====================================================
# üíæ SAVE MODEL HERE
# =====================================================

model.save("rbfnn_model.h5", save_format="h5")
print("üíæ RBFNN Model saved as: rbfnn_model.h5")

# =====================================================
#                 EVALUATION
# =====================================================

y_probs = model.predict(X_test).flatten()
y_pred = (y_probs > 0.5).astype(int)

print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))
print("üßæ Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# ‚úÖ Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_curve, auc
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Dropout, Conv1D, MaxPooling1D, Flatten, SimpleRNN, LSTM
from tensorflow.keras.callbacks import ReduceLROnPlateau
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import MinMaxScaler
from imblearn.combine import SMOTEENN
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# =====================================================
#                 LOAD & PREPARE DATA
# =====================================================

df = pd.read_csv('./24-04-2025-1192ALLSAMPLES.csv')
X = df.drop('class', axis=1)
y = df['class']

# Feature selection
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
top_features = pd.Series(rf.feature_importances_, index=X.columns).nlargest(10).index.tolist()
X = df[top_features]

# Normalize
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# SMOTEENN combination
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X_scaled, y)

# =====================================================
#                 GAN SYNTHETIC DATA
# =====================================================

class GANGen(nn.Module):
    def __init__(self, z_dim, x_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(z_dim, 64), nn.ReLU(),
            nn.Linear(64, x_dim), nn.Sigmoid()
        )
    def forward(self, z): return self.model(z)

class GANDisc(nn.Module):
    def __init__(self, x_dim):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(x_dim, 64), nn.ReLU(),
            nn.Linear(64, 1), nn.Sigmoid()
        )
    def forward(self, x): return self.model(x)

def generate_gan(X_class, n_samples=500, z_dim=32, epochs=100):
    device = torch.device("cpu")
    G = GANGen(z_dim, X_class.shape[1]).to(device)
    D = GANDisc(X_class.shape[1]).to(device)
    g_opt = optim.Adam(G.parameters(), lr=0.0002)
    d_opt = optim.Adam(D.parameters(), lr=0.0002)
    loss_fn = nn.BCELoss()

    loader = DataLoader(TensorDataset(torch.tensor(X_class, dtype=torch.float32)), batch_size=64, shuffle=True)

    for _ in range(epochs):
        for real_batch, in loader:
            real_batch = real_batch.to(device)
            bs = real_batch.size(0)
            real_labels = torch.ones(bs, 1).to(device)
            fake_labels = torch.zeros(bs, 1).to(device)
            z = torch.randn(bs, z_dim).to(device)
            fake_batch = G(z)

            d_real = D(real_batch)
            d_fake = D(fake_batch.detach())
            d_loss = loss_fn(d_real, real_labels) + loss_fn(d_fake, fake_labels)
            d_opt.zero_grad(); d_loss.backward(); d_opt.step()

            g_loss = loss_fn(D(fake_batch), real_labels)
            g_opt.zero_grad(); g_loss.backward(); g_opt.step()

    with torch.no_grad():
        z = torch.randn(n_samples, z_dim).to(device)
        return G(z).cpu().numpy()

# Generate GAN samples
X_class1 = X_resampled[y_resampled == 1]
X_class0 = X_resampled[y_resampled == 0]

gan_data_1 = generate_gan(X_class1, n_samples=500)
gan_data_0 = generate_gan(X_class0, n_samples=500)

# Combine data
X_final = np.vstack([X_resampled, gan_data_1, gan_data_0])
y_final = np.concatenate([y_resampled,
                          np.ones(len(gan_data_1)),
                          np.zeros(len(gan_data_0))]).astype(int)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_final, y_final, test_size=0.2, stratify=y_final, random_state=42
)

# Reshape for CNN / RNN / LSTM
X_train_seq = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test_seq = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))


# =====================================================
#                 MODEL DEFINITIONS
# =====================================================

def build_cnn(input_shape):
    model = Sequential([
        Conv1D(32, 3, activation='relu', input_shape=input_shape),
        MaxPooling1D(2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_rnn(input_shape):
    model = Sequential([
        SimpleRNN(64, input_shape=input_shape),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def build_lstm(input_shape):
    model = Sequential([
        LSTM(64, input_shape=input_shape),
        Dropout(0.3),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model


# =====================================================
#     TRAIN, EVALUATE, SAVE MODEL FUNCTION
# =====================================================

def train_and_evaluate(model_fn, name):
    model = model_fn((X_train_seq.shape[1], 1))

    history = model.fit(
        X_train_seq, y_train,
        epochs=100,
        batch_size=16,
        validation_split=0.1,
        verbose=1,
        callbacks=[ReduceLROnPlateau(monitor='val_loss', patience=3)]
    )

    # ===========================
    # üíæ SAVE TRAINED MODEL HERE
    # ===========================
    save_path = f"{name}_model.h5"
    model.save(save_path)
    print(f"üíæ Saved {name} model as: {save_path}")
    # ===========================

    y_probs = model.predict(X_test_seq).flatten()
    y_pred = (y_probs > 0.5).astype(int)

    print(f"\nüìå {name} Classification Report:\n")
    print(classification_report(y_test, y_pred))
    print("üìå Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("‚úÖ Accuracy:", accuracy_score(y_test, y_pred))

    # Save prediction table
    result_df = pd.DataFrame({
        'True Label': y_test,
        'Predicted Label': y_pred,
        'Confidence Score': y_probs
    })
    result_df.index.name = 'Sample Index'
    result_df.to_csv(f'{name}_Predictions.csv', index=True)

    print("\nüîç First 10 Predictions:")
    print(result_df.head(10))

    return result_df

# =====================================================
#                 RUN & SAVE ALL MODELS
# =====================================================

cnn_results = train_and_evaluate(build_cnn, "CNN")
rnn_results = train_and_evaluate(build_rnn, "SimpleRNN")
lstm_results = train_and_evaluate(build_lstm, "LSTM")
```
